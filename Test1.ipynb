```python
from selenium import webdriver
import time
import os

def capture_screenshot(driver, link, device, resolution, timestamp):
    screenshot_folder = os.path.join("Screenshots", device, resolution)
    os.makedirs(screenshot_folder, exist_ok=True)

    screenshot_path = os.path.join(screenshot_folder, f"Screenshot-{timestamp}-{link_hash(link)}.png")
    driver.save_screenshot(screenshot_path)

def link_hash(link):
    # A simple hash function for creating unique filenames based on the link
    return str(hash(link))

def run_test(link, device, resolution, browser="chrome"):
    timestamp = time.strftime("%Y%m%d-%H%M%S")

    if browser.lower() == "chrome":
        driver = webdriver.Chrome()
    elif browser.lower() == "firefox":
        driver = webdriver.Firefox()
    else:
        raise ValueError(f"Unsupported browser: {browser}")

    try:
        # Navigate to the provided link
        driver.get(link)

       
        # Capture a screenshot
        capture_screenshot(driver, link, device, resolution, timestamp)

    except Exception as e:
        print(f"Error during test for {link}: {e}")

    finally:
        # Close the browser window
        driver.quit()

if __name__ == "__main__":
    devices = {"Desktop": ["1920x1080", "1366x768", "1536x864"],
               "Mobile": ["360x640", "414x896", "375x667"]}

    #top 5  links
    list_links = [
        "https://www.getcalley.com/calley-call-from-browser/",
        "https://www.getcalley.com/calley-pro-features/",
        "https://www.getcalley.com/best-auto-dialer-app/",
        "https://www.getcalley.com/how-calley-auto-dialer-app-works/",
        "https://example.com"
    ]

    for device, resolutions in devices.items():
        for resolution in resolutions:
            for link in list_links:
                run_test(link, device, resolution, browser="chrome")
                run_test(link, device, resolution, browser="firefox")

                
```
